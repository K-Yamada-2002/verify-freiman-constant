```python
alphas = []

#type1 -> type1 + type2
alphas.append(continued_fraction([(1,),(4,1)]))
alphas.append(continued_fraction([(1,),(1,4)]))
alphas.append(continued_fraction([(2,),(4,1)]))
alphas.append(continued_fraction([(4,),(1,4)]))

#type2 -> type1 + type3
alphas.append(continued_fraction([(2,),(4,1)]))
alphas.append(continued_fraction([(2,),(1,4)]))
alphas.append(continued_fraction([(3,),(4,1)]))
alphas.append(continued_fraction([(4,),(1,4)]))

#type3 -> type1 + type1
alphas.append(continued_fraction([(3,),(4,1)]))
alphas.append(continued_fraction([(3,),(1,4)]))
alphas.append(continued_fraction([(4,),(4,1)]))
alphas.append(continued_fraction([(4,),(1,4)]))

for i in range(3):
    a_0 = alphas[i * 4 + 0].value()
    a_1 = alphas[i * 4 + 1].value()
    a_2 = alphas[i * 4 + 2].value()
    a_3 = alphas[i * 4 + 3].value()

    print("type {}:".format(i + 1))
    #M1/I の下からの評価
    M1_to_I_bound = abs((a_0 - a_1) / (a_2 - a_1)) \
                    * (a_2 + 1) / (a_0 + 1)
    print("|M1| / |I| ≧ {}".format(n(M1_to_I_bound)))
    
    #M2/I の下からの評価
    M2_to_I_bound = abs((a_2 - a_3) / (a_2 - a_1)) \
                    * a_1 / a_3
    print("|M2| / |I| ≧ {}".format(n(M2_to_I_bound)))
```
