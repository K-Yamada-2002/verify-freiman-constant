```python
def ratio_of_intervals(alpha, beta, gamma, delta) :
    # let A be an interval [0;a_1,...,a_n,alpha] ~ [0;a_1,...,a_n,beta] and
    # let B be an interval [0;a_1,...,a_n,gamma] ~ [0;a_1,...,a_n,delta].
    # This functon returns the lower and upper bounds of |A|/|B|

    # |A| / |B| = abs(alpha - beta) / abs(gamma - delta) * 
    #    (gamma + q_{n-1}/q_n)(delta * q_{n-1}/q_n) / (alpha + q_{n-1}/q_n)(beta +q_{n-1}/q_n).
    # t = q_{n-1} / q_n 
    f(t) = abs(alpha - beta) / abs(gamma - delta) * (gamma + t) * (delta + t) / ((alpha + t) * (beta + t))

    # q_{n-1} / q_n  varies 0 to 1. So candidate of lower and upper bounds of |A| / |B| is f(0), f(1), and f(t) s.t. f'(t) = 0
    candidate = [f(0), f(1)]
    critical_points = solve(f.diff(t) == 0, t)
    for critical_point in critical_points:
        if 0 < critical_point.rhs() and critical_point.rhs() < 1:
            candidate.append(f(critical_point.rhs()))
            
    return (n(min(candidate)),n(max(candidate)))

delta_1 = continued_fraction([(0,),(1,3)]).value()
delta_2 = continued_fraction([(0,4),(1,3)]).value()
nu = continued_fraction([(0,),(3,1)]).value()

print("--- type1 -> type2 + type2 + type1 + type1 ---")
original_interval = [1 + nu, 4 + delta_1]
removed_intervals = []

removed_intervals.append((1 + delta_1, 2 + nu))
removed_intervals.append((2 + delta_1, 3 + delta_2))
removed_intervals.append((3 + delta_1, 4 + delta_2))

for removed in removed_intervals:
    print(ratio_of_intervals(original_interval[0], removed[0], removed[0], removed[1])[0])
    print(ratio_of_intervals(removed[1], original_interval[1], removed[0], removed[1])[0])
    original_interval[0] = removed[1]

print("--- type2 -> type2 + type2 + type1 ---")
original_interval = [1 + nu, 3 + delta_1]
removed_intervals = []

removed_intervals.append((1 + delta_1, 2 + nu))
removed_intervals.append((2 + delta_1, 3 + delta_2))

for removed in removed_intervals:
    print(ratio_of_intervals(original_interval[0], removed[0], removed[0], removed[1])[0])
    print(ratio_of_intervals(removed[1], original_interval[1], removed[0], removed[1])[0])
    original_interval[0] = removed[1]
```
